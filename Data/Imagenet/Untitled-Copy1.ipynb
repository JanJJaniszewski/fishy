```python
import glob
import urllib.request
import sys
```

```python
for foldername in glob.glob('./*'):
    for filename in glob.glob(foldername + '/*'):
        f = open(filename, 'r')
        for url in f:
            try:
                pic = open(foldername + '/' + url.split('/')[-1][:-1],'wb')
                pic.write(urllib.request.urlopen(url).read())
                pic.close()
            except:
                print("Unexpected error:", sys.exc_info())
        f.close()
```
